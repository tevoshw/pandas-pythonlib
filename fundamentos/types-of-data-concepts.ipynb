# TYPE OF DATAS

### CONCEPTS OF DATA, PANDAS, MACHINE LEARNING AND DATA SCIENCE
- DATA: ANY INFORMATION ("ana", 17, "email spam")
- DATA TYPE or DTYPE: NATURE OF DATA (int, floating, booelan, text, binary, time/hour and etc)
- DATABASE: PERMANENT STRUCTURED DATA, JUST FOR STORAGE THE DATA (tables, mysql, and etc)
- DATASET: A PART OF, OR THE DATA, SELECT FOR THE USE IN THE DATAFRAMES
- DATAFRAME: DATA FOR ANALISE AND VISUALIZATION 

- SAMPLE: AN EXAMPLE OF DATA, WITH ALL THE COLUMNS (HORIZONTAL)
- FEATURES: A ROW, INDEPENDENTS DATA
- LABEL: THE TARGET OF THE PREDICTION, NORMALLY A ROW, DEPENDENS DATA (FEATURE)
- SERIES: A FEATURE WITH ALL THE INDEXS (VERTICAL)
- INDEX: DATA ORGANIZE IN 0, 1, 2, 3 AND ETC

### STRUCTURED DATA
- WITH COLUMNS AND ROWS
- TABLE, DATABASE AND MEANING



### UNSTRUCTURED DATA
- WITH NO COLUMNS AND ROWS
- NO TABLE, DATABASE AND MEANING
