# INDEXAÇÃO BOOLEANA É UMA FORMA DE ANALISAR E VISUALIZAR OS DADOS BASEADOS EM TIPOS BOOLEANOS E ESTRUTURAS DE CONTROLE -> return samples

```python
import pandas as pd
df = pd.read_csv('../datasets/GasPricesinBrazil_2004-2019.csv', sep = ';')
```

```python
df[df["MÊS"] > 10] # RETURN THE SAMPLE THAT CONDITION WE GIVE
```

```python
df[ (df["MÊS"] > 10) & (df["PREÇO MÉDIO REVENDA"] > 95) ] # & MEABS AD
```

```python
df[ (df["DESVIO PADRÃO REVENDA"] > 10) | (df["PREÇO MÉDIO REVENDA"] > 95) ] # | MEANS OR
```
