```python
import pandas as pd
df = pd.read_csv('../datasets/GasPricesinBrazil_2004-2019.csv', sep = ';')
df.head()
```

##### .LOC[]

```python
df.loc[0] # THE FIRST SAMPLE
# RETORNA A 1 LINHA // O 1 SAMPLE 
# BOOLEAN CONDITIONS
# df.loc[row, features]
```

```python
df.loc[ [0,1] ] # MORE THAN ONE SAMPLE
```

```python
# using the df.loc[row, features]

df.loc[ df["NÚMERO DE POSTOS PESQUISADOS"] > 300, "REGIÃO"] # ONE FEATURE ONLY
```

```python
df.loc[ df["NÚMERO DE POSTOS PESQUISADOS"] > 300, ["REGIÃO", "MÊS"]] # MORE THAN ONE FEATURE
```
